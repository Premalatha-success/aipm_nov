```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
pd.set_option('display.max_columns', None)
data = pd.read_csv("/content/CardioGoodFitness-1.csv")
```

```python
Male = data['Gender'] == 'Male'
maledata= data[Male]
Female = data['Gender'] == 'Female'
femaledata= data[Female]
single = data['MaritalStatus'] == 'Single'
singledata= data[single]
partnered = data['MaritalStatus'] == 'Partnered'
partnereddata= data[partnered]
```

```python
plt.figure(figsize=(15, 7)) #change figure size
sns.countplot(x="Gender", hue= 'Product', data= data) #creates count plot for the number of male and female customers that bought the different products.
plt.title("Number of male and female customers with respect to product bought", fontsize = 30) #titles the graph.
plt.ylabel('Product bought', fontsize = 20) #changes y axis label.
plt.xlabel('Gender', fontsize = 20) #changes x axis label.
plt.show() #shows graph.
```
